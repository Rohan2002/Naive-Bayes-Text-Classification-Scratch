```python
import pandas as pd
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize 
from nltk.corpus import stopwords
import nltk
import string
from sklearn.feature_extraction.text import CountVectorizer
from nltk.probability import FreqDist
```

```python

# True:1 => answer, False:0 => "question, Just try "what" type questions

columns = ['text', 'label']
rows = []
# True:1 => answer, False:0 => "question
rows = [['This is my book', 'answer'], 
        ['They are novels', 'answer'],
        ['have you read this book', 'question'],
        ['who is the author', 'question'],
        ['what are the characters', 'question'],
        ['This is how I bought the book', 'answer'],
        ['I like fictions', 'answer'],
        ['what is your favorite book', 'question']]

df = pd.DataFrame(rows, columns=columns)
nltk.download('punkt')
nltk.download('stopwords')

N=len(df.index) # DEFAULT GLOBAL VARIABLE FOR THE ROW SIZE
true_arr =[]
false_arr = []

def clearStopWords(aList):
    stop_words = set(stopwords.words('english')) 
    filtered_sentence = [] 
    for w in aList: 
        if w not in stop_words: 
            filtered_sentence.append(w) 
    return filtered_sentence
    
def cleanSentence(sentence):
    lowerSent = sentence.lower()
    punc = string.punctuation + "‘’"
    NoPunc = lowerSent.translate(str.maketrans('', '', punc))
    NoDigits = NoPunc.translate(str.maketrans('','',string.digits))
    return NoDigits

#here
def ListText():
    true_text = ''
    false_text = ''
    for row in df.head(N).itertuples():
        if row.label == 'answer':
            true_text+= cleanSentence(row.text) + " "
        else:
            false_text+=cleanSentence(row.text) + " "
        
        true_arr = clearStopWords(word_tokenize(true_text))
        false_arr = clearStopWords(word_tokenize(false_text))
    return (true_arr, false_arr)

def classTrueProbability():
    trueClassSum = 0
    for row in df.itertuples():
        if row.label == 'answer':
            trueClassSum +=1
    return trueClassSum/N 
print(classTrueProbability())

def classFalseProbability():
    return 1 - classTrueProbability()
# print(classFalseProbability())

def frequencyTrue():
    ListWords = FreqDist(ListText()[0])
    freq_q = {}
    for word,value in ListWords.items():
        freq_q[word] = value
    return freq_q
# print(frequencyTrue())

def frequencyFalse():
    ListWords = FreqDist(ListText()[1])
    freq_q = {}
    for word,value in ListWords.items():
        freq_q[word] = value
    return freq_q
# print(frequencyFalse())

def ProbabilityTrue():
    freq_true = frequencyTrue()
    for word,value in freq_true.items():
        freq_true[word] = value/len(freq_true)
    return freq_true
# print(ProbabilityTrue())

def ProbabilityFalse():
    freq_False = frequencyFalse()
    for word,value in freq_False.items():
        freq_False[word] = value/len(freq_False)
    return freq_False
# print(ProbabilityFalse())

def sumTotalFeatures():
    docs = [row.text for row in df.head(N).itertuples()]
    vec = CountVectorizer()
    X = vec.fit_transform(docs)
    total_features = len(vec.get_feature_names())
    return total_features
# print(sumTotalFeatures())
def sumFeaturesTrue():
    freq_true = frequencyTrue()
    sumTrue =0
    for word,value in freq_true.items():
        sumTrue+=value
    return sumTrue
# print(sumFeaturesTrue())

def sumFeaturesFalse():
    freq_False = frequencyFalse()
    sumFalse =0
    for word,value in freq_False.items():
        sumFalse+=value
    return sumFalse
# print(sumFeaturesFalse())

# Add Laplace Smoothing to handle a missing word from the training set
def LaplaceTrueProbability():
    LapLaceTrueTemp = dict.fromkeys(word_tokenize("what is the answer"),0) # Change ur sentence here 
    FrequencyTrue = frequencyTrue()
    for word in LapLaceTrueTemp:
        if word in FrequencyTrue.keys():
            count = FrequencyTrue[word]
        else:
            count = 0
        LapLaceTrueTemp[word] = (count + 1)/(sumFeaturesTrue() + sumTotalFeatures())
    return LapLaceTrueTemp
# print(LaplaceTrueProbability())

def LaplaceFalseProbability():
    LapLaceFalseTemp = dict.fromkeys(word_tokenize("what is the answer"),0)  # Change ur sentence here 
    FrequencyFalse = frequencyFalse()
    for word in LapLaceFalseTemp:
        if word in FrequencyFalse.keys():
            count = FrequencyFalse[word]
        else:
            count = 0
        LapLaceFalseTemp[word] = (count + 1)/(sumFeaturesFalse() + sumTotalFeatures())
    return LapLaceFalseTemp
# print(LaplaceFalseProbability())

def FinalDecision():
    trueprob = LaplaceTrueProbability()
    falseprob = LaplaceFalseProbability()
    multiplier_t = 1
    multiplier_f = 1
    for word_t in trueprob:
        multiplier_t*=trueprob[word_t]
    
    for word_f in falseprob:
        multiplier_f*=falseprob[word_f]
    
    print(multiplier_f, multiplier_t)
    if (multiplier_t * classTrueProbability()) > (multiplier_f * classFalseProbability()):
        return "answer"
    if (multiplier_t * classTrueProbability()) < (multiplier_f* classFalseProbability() ):
        return "question"
    

print(FinalDecision())
```
